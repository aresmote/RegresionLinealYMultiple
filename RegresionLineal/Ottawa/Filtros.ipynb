```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('Datos_Limpios_OTTAWA.csv')
data.info
```

**CONVERSIÃ“N EN LOS TIPOS DE DATOS**

**PRICE**

```python
data['price'] = data['price'].str.replace(r'[^0-9.-]', '', regex=True)
data['price'] = data['price'].astype(float)

data['price'].info()
```

```python
data['price'].head(10)
```

**HOST_IS_SUPERHOST**

```python
data['host_is_superhost'] = data['host_is_superhost'].replace({'t': '1'}, regex=False)
data['host_is_superhost'] = data['host_is_superhost'].replace({'f': '0'}, regex=False)
data['host_is_superhost'].head(10)
```

```python
data['host_is_superhost'] = data['host_is_superhost'].astype(float)

data['host_is_superhost'].info()

data['host_is_superhost'].head(5)
```
